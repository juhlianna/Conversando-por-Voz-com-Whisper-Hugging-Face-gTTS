2. Importar bibliotecas

```python
import whisper
from transformers import pipeline
from gtts import gTTS
from IPython.display import Audio
```

3. Transcrição de áudio com Whisper
Faça upload de um arquivo de áudio (fala.mp3) no Colab e rode:

```python
# Carregar modelo Whisper (pode usar "base", "small", "medium", "large")
model = whisper.load_model("small")
result = model.transcribe("fala.mp3")
print("Transcrição:", result["text"])

```

4. Resposta com modelo Hugging Face

```python
# Modelo em português
chat = pipeline("text2text-generation", model="unicamp-dl/ptt5-base-portuguese-vocab")
resposta = chat(result["text"], max_new_tokens=50)[0]["generated_text"]
print("Resposta:", resposta)



```

5. Converter resposta em áudio (gTTS)

```python
tts = gTTS(resposta, lang="pt")
tts.save("resposta.mp3")
Audio("resposta.mp3")

```
