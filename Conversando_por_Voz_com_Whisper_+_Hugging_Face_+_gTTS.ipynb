2. Importar bibliotecas

```python
import whisper
from transformers import pipeline
from gtts import gTTS
from IPython.display import Audio
```

3. Transcrição de áudio com Whisper
Faça upload de um arquivo de áudio (fala.mp3) no Colab e rode:

```python
# Carregar modelo Whisper (pode usar "base", "small", "medium", "large")
model = whisper.load_model("small")

def transcrever_audio(caminho_arquivo):
    result = model.transcribe(caminho_arquivo)
    return result["text"]
```

4. Resposta com modelo Hugging Face

```python
# Carregar modelo de linguagem open source
chat = pipeline("text-generation", model="mistralai/Mistral-7B-Instruct-v0.2")

def conversar(texto_usuario):
    resposta = chat(texto_usuario, max_new_tokens=100)[0]["generated_text"]
    return resposta
```

5. Converter resposta em áudio (gTTS)

```python
def resposta_em_audio(texto, arquivo_saida="resposta.mp3"):
    tts = gTTS(texto, lang="pt")
    tts.save(arquivo_saida)
    return Audio(arquivo_saida, autoplay=True)
```

6. Fluxo completo: áudio → texto → resposta → voz

```python
def chat_por_voz(caminho_audio):
    # 1. Transcreve o áudio
    texto_usuario = transcrever_audio(caminho_audio)
    print("🎤 Você disse:", texto_usuario)

    # 2. Modelo responde
    resposta = conversar(texto_usuario)
    print("🤖 Modelo respondeu:", resposta)

    # 3. Converte resposta em áudio
    return resposta_em_audio(resposta)
```

7. Testando
Depois de fazer upload de um arquivo de áudio (ex.: fala.mp3), rode:

```python
chat_por_voz("fala.mp3")
```
